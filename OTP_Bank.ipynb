```python
# заменяем запятые на точки и преобразуем в тип float
df_otp['FST_PAYMENT'] = df_otp['FST_PAYMENT'].str.replace(',', '.').astype('float', copy = False)
df_otp['PERSONAL_INCOME'] = df_otp['PERSONAL_INCOME'].str.replace(',', '.').astype('float', copy = False)
df_otp['CREDIT'] = df_otp['CREDIT'].str.replace(',', '.').astype('float', copy = False)
df_otp['LOAN_AVG_DLQ_AMT'] = df_otp['LOAN_AVG_DLQ_AMT'].str.replace(',', '.').astype('float', copy = False)
df_otp['LOAN_MAX_DLQ_AMT'] = df_otp['LOAN_MAX_DLQ_AMT'].str.replace(',', '.').astype('float', copy = False)
```

```python
k = ['EDUCATION', 'MARITAL_STATUS', 'GEN_INDUSTRY','GEN_TITLE', 'ORG_TP_STATE', 'JOB_DIR', 'FAMILY_INCOME', 'REG_ADDRESS_PROVINCE', 'FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE', 'TP_PROVINCE']
for i in k:
    print(df_otp[i].value_counts())
```

```python
# Избавляемся от редких категорий
df_otp['EDUCATION'].replace(['Два и более высших образования', 'Ученая степень'],'Среднее специальное' , inplace = True)
df_otp['GEN_INDUSTRY'].replace(['Маркетинг', 'Подбор персонала', 'Логистика','Управляющая компания'], 'Торговля' , inplace = True)
df_otp['GEN_TITLE'].replace('Партнер', 'Специалист' , inplace = True)
df_otp['ORG_TP_STATE'].replace('Частная ком. с инос. капиталом', 'Частная компания' , inplace = True)
df_otp['JOB_DIR'].replace(['Реклама и маркетинг', 'Юридическая служба'], 'Участие в основ. деятельности' , inplace = True)
df_otp['REG_ADDRESS_PROVINCE'].replace(['Дагестан','Агинский Бурятский АО','Усть-Ордынский Бурятский АО','Эвенкийский АО'], 'Кемеровская область' , inplace = True)
df_otp['FACT_ADDRESS_PROVINCE'].replace(['Агинский Бурятский АО', 'Эвенкийский АО', 'Усть-Ордынский Бурятский АО','Дагестан'], 'Краснодарский край' , inplace = True)
df_otp['POSTAL_ADDRESS_PROVINCE'].replace(['Усть-Ордынский Бурятский АО','Агинский Бурятский АО', 'Эвенкийский АО'], 'Краснодарский край' , inplace = True)
df_otp['TP_PROVINCE'].replace(['Москва', 'Кабардино-Балкария'], 'Краснодарский край', inplace = True)
```

```python
# избавляемся от пропусков, где они есть
df_otp.loc[:, df_otp.isnull().any()] # cтолбцы, где есть пропуски
```

```python
# разбиваем выборку на train и test
train = df_otp.sample(frac = 0.7, random_state = 200)
test = df_otp.drop(train.index)
```

```python
df_otp.loc[:, df_otp.isnull().any()] # Выводим столбцы, в которых есть пропуски
```

```python
m =['GEN_TITLE', 'ORG_TP_STATE','GEN_INDUSTRY','ORG_TP_FCAPITAL', 'JOB_DIR', 'TP_PROVINCE', 'REGION_NM', 'WORK_TIME'] 
for i in m:
    print(i, df_otp[i].value_counts())
```

```python
# Заполняем пропуски в предикторах
test['REGION_NM'].fillna('ЮЖНЫЙ', inplace = True)
train['REGION_NM'].fillna('ЮЖНЫЙ', inplace = True)

train['GEN_INDUSTRY'].fillna('Другие сферы', inplace = True)
test['GEN_INDUSTRY'].fillna('Другие сферы', inplace = True)

train['GEN_TITLE'].fillna('Другое', inplace = True)
test['GEN_TITLE'].fillna('Другое', inplace = True)

train['ORG_TP_STATE'].fillna('Нет', inplace = True)
test['ORG_TP_STATE'].fillna('Нет', inplace = True)

train['ORG_TP_FCAPITAL'].fillna('Нет', inplace = True)
test['ORG_TP_FCAPITAL'].fillna('Нет', inplace = True)

train['JOB_DIR'].fillna('Безработный', inplace = True)
test['JOB_DIR'].fillna('Безработный', inplace = True)

train['TP_PROVINCE'].fillna('Не брал кредит', inplace = True)
test['TP_PROVINCE'].fillna('Не брал кредит', inplace = True)

train['WORK_TIME'].fillna(0, inplace = True)
test['WORK_TIME'].fillna(0, inplace = True)

train['PREVIOUS_CARD_NUM_UTILIZED'].fillna(0, inplace = True)
test['PREVIOUS_CARD_NUM_UTILIZED'].fillna(0, inplace = True)
```

```python
train.info()
```

```python
# Вновь посмотрим на наши предикторы. Видно, что в некоторых признаках очень много категорий. 
# Попробуем сделать их количественными, применив Frequency Encoding

for i in train.columns:
    print(i, pd.unique(train[i]))
```

```python
# заменим категориальные предикторы численными значениями, равными их частоте в тестовой и контрольной выборке, а потом сравним с dummy-кодированием.
k = ['GEN_INDUSTRY', 'GEN_TITLE', 'REG_ADDRESS_PROVINCE', 'FACT_ADDRESS_PROVINCE', 'POSTAL_ADDRESS_PROVINCE', 'TP_PROVINCE']
for i in k:
    for j in list(train[i].unique()):
        train[i].replace(j, train[train[i] == j][i].count()/train[i].value_counts().sum() , inplace = True)
```

```python
# аналогично для теста
for i in k:
    for j in list(test[i].unique()):
        test[i].replace(j, test[test[i] == j][i].count()/test[i].value_counts().sum() , inplace = True)
```

```python
train.head(100)
# Можно попробовать добавить следующие признаки:
# FAMILY_INCOME/PERSONAL_INCOME - если данный признак стремится к 1, дефолтность должна увеличиться. если к бесконечности, то уменьшиться
# LOAN_NUM_TOTAL/LOAN_NUM_CLOSED - если стремится к 1, то все хорошо. если к бесконечности, тто плохо.
# CREDIT/TERM - ежемес. платеж
```

```python
train['Tot/Clos'] = train['LOAN_NUM_CLOSED']/train['LOAN_NUM_TOTAL']
train['Month_amount'] = train['CREDIT']/train['TERM']
test['Tot/Clos'] = test['LOAN_NUM_CLOSED']/test['LOAN_NUM_TOTAL']
test['Month_amount'] = test['CREDIT']/test['TERM']
```

```python
# Теперь выполним dummy-кодирование остальных категориальных признаков, количество категорий в которых не такое большое.
train_dummy = pd.get_dummies(train)
test_dummy = pd.get_dummies(test)
```

```python
# создаем обучающий и контрольный массивы меток
y_train = train_dummy.loc[:, 'TARGET']
y_test = test_dummy.loc[:, 'TARGET']
# создаем обучающий и контрольный массивы признаков
X_train = train_dummy.drop('TARGET', axis = 1)
X_test = test_dummy.drop('TARGET', axis = 1)
```

```python
# Для начала попробуем обучить одно дерево
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth = 6)
tree.fit(X_train, y_train)
print('Правильность на обучающей выборке: {:.3f}'.format(tree.score(X_train, y_train)))
print('Правильность на контрольной выборке: {:.3f}'.format(tree.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 500, max_depth = 10, max_features = 50, min_samples_split = 2, min_samples_leaf = 1, random_state=152)
forest.fit(X_train, y_train)
print('Правильность на обучающей выборке: {:.3f}'.format(forest.score(X_train, y_train)))
print('Правильность на контрольной выборке: {:.3f}'.format(forest.score(X_test, y_test)))
```

```python
# Качество AUC для дерева близко к случайному классификатору
from sklearn.metrics import roc_auc_score
print('AUC на обучающей выборке: {:.3f}'.
      format(roc_auc_score(y_train, tree.predict_proba(X_train)[:,1])))
print('AUC на контрольной выборке: {:.3f}'.
      format(roc_auc_score(y_test, tree.predict_proba(X_test)[:, 1])))
```

```python
from sklearn.metrics import roc_auc_score
print('AUC на обучающей выборке: {:.3f}'.
      format(roc_auc_score(y_train, forest.predict_proba(X_train)[:,1])))
print('AUC на контрольной выборке: {:.3f}'.
      format(roc_auc_score(y_test, forest.predict_proba(X_test)[:, 1])))
```

```python
best_score = 0
for max_depth in [4, 6, 8, 10, 12, 14, 16, 18]:
    for max_features in [4,8,16,32,64]:
        # для каждой комбинации параметров обучаем дерево CART
            forestt = RandomForestClassifier(max_depth=max_depth, max_features=max_features, n_estimators = 500, random_state=42)
            forestt.fit(X_train, y_train)
        # оцениваем качество дерева CART на контрольной выборке
            auc_score = roc_auc_score(y_test, forestt.predict_proba(X_test)[:, 1])
        # если получаем наилучшее значение AUC, сохраняем значение и параметры
            if auc_score > best_score:
                best_score = auc_score
                best_parameters = {'max_features': max_features, 'max_depth': max_depth}

print('Наилучшее значение AUC: {:.2f}'.format(best_score))
print('Наилучшие значения параметров: {}'.format(best_parameters))
```
