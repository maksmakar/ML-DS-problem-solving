```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:/Datasets/SeverStal/Stats in 2016.csv', sep = ',', encoding = 'cp1251')
```

```python
df
```

```python
# Выведем частоты количества больничных дней больше 2 для мужчин и женщин как M и W соответственно
print('Мужчины:\n {}'.format(df[(df['Пол'] == 'М') & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()))
print('Женщины:\n {}'.format(df[(df['Пол'] == 'Ж') & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()))
```

```python
# Создадим списки из значений частот по количеству больничных больше 2 дней для мужчин и женщин
M_1 = [i for i in df[(df['Пол'] == 'М') & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()]
W_1 = [i for i in df[(df['Пол'] == 'Ж') & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()]
```

```python
# Вычисляем критерий хи-квадрат
scipy.stats.mstats.chisquare(M_1, f_exp = W_1) 
```

```python
# Произведем аналогичные расчеты для сотрудников старше 35 и младше 35 лет
print('Работники старше 35 лет:\n {}'.format(df[(df['Возраст'] > 35) & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()))
print('Работники старше 35 лет:\n {}'.format(df[(df['Возраст'] <= 35) & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()))
```

```python
M_2 = [i for i in df[(df['Возраст'] > 35) & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()]
W_2 = [i for i in df[(df['Возраст'] <= 35) & (df['Количество больничных дней'] > 2)]\
                             ['Количество больничных дней'].value_counts().sort_index()]
```

```python
scipy.stats.mstats.chisquare(M_2, f_exp = W_2) 
```
